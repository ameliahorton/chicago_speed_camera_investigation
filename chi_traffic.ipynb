```python
# imports
import pandas as pd
import datetime
```

```python
# read in csv file to dataframe
original_df = pd.read_csv('Speed_Camera_Violations1.csv')
original_df
```

```python
# drop null values
df = original_df.dropna()
df
```

```python
# convert date column to datetime
df['VIOLATION DATE'] =pd.to_datetime(df['VIOLATION DATE'])
df 
```

```python
# run value_counts() on years
df['VIOLATION DATE'].dt.year.value_counts()
```

```python
# remove data from 2014
filtered_df = df.loc[df['VIOLATION DATE'] > '2014-12-31']
filtered_df
```

```python
daily_violations = filtered_df.sort_values('VIOLATIONS', ascending=False)
daily_violations
```

```python
total_violations = filtered_df['VIOLATIONS'].sum()
total_violations
```

```python
address_list = []

for i in range(len(filtered_df)):
    address = filtered_df.iloc[i,1]
    if address not in address_list:
        address_list.append(address)

len(address_list)
```

```python
# create a dataframe that lists each unique address, the total number of violations at that address in the entire dataset, and the lat and lng location of the address
grouped_df = filtered_df.groupby(['ADDRESS'], as_index=False, sort=False).agg({'VIOLATIONS':'sum', 'LATITUDE':'first', 'LONGITUDE':'first'})
sorted_and_grouped = grouped_df.sort_values('VIOLATIONS', ascending=False)
sorted_and_grouped.head(20)
```

```python
# export sorted_and_grouped dataframe to json file
json = sorted_and_grouped.to_json('./speed_export.json', orient='index')
```
